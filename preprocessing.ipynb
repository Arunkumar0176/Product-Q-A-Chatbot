# With description product list

```python
files2 = ['/content/flipcartHeadphones (1).csv', '/content/cosmetic_Product (1).csv', '/content/flipcartTV.csv']
df_list2 = []
columns_map = {
    'productName': 'productName',
    'description': 'description',
    'productDescription': 'description',
    'question': 'question',
    'answer': 'answer'
}
for file in files2:
    temp = pd.read_csv(file)
    print(temp.columns)
    temp.rename(columns=columns_map, inplace=True)
    df = temp[['productName','description', 'question', 'answer']]
    df_list.append(df)
final_df2 = pd.concat(df_list, ignore_index=True)
print(final_df2.columns, final_df2.shape)
```

```python
print(final_df2.head())
final_df2.dropna(inplace=True)
final_df2.reset_index(drop=True, inplace=True)
print(final_df2.columns, final_df2.shape)
```

## Converting into hugging face datasetDict format

```python
final_df2.to_csv("my_data.csv", index=False)

# Download the file
from google.colab import files
files.download("my_data.csv")
```

```python
data_dictionary = Dataset.from_pandas(final_df2).train_test_split(test_size=0.2)
print(data_dictionary)
```

```python
data_dictionary.save_to_disk("my_local_datasetDictionary")

print("Dataset saved successfully!")
```
